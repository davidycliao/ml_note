```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
from math import e
```

```python
n_data = torch.ones(100, 2)
x0 = torch.normal(2*n_data, 1)      # class0 x data (tensor), shape=(100, 2)
y0 = torch.zeros(100)               # class0 y data (tensor), shape=(100, 1)
x1 = torch.normal(-2*n_data, 1)     # class1 x data (tensor), shape=(100, 2)
y1 = torch.ones(100)                # class1 y data (tensor), shape=(100, 1)
x = torch.cat((x0, x1), 0).type(torch.FloatTensor)  # shape (200, 2) FloatTensor = 32-bit floating
y = torch.cat((y0, y1),  ).type(torch.LongTensor)
```

```python
x.size()
```

```python
y.size()
```

```python
print('requires_grad for X_train: ', x.requires_grad)
print('requires_grad for Y_train: ', y.requires_grad)
```

```python
input_size = 2 
hidden_size = 10
output_size = 2 
learning_rate = 1e-6
```

```python
w1 = torch.rand(input_size, 
                hidden_size, 
                
                requires_grad=True)
w1.shape
```

```python
w1
```

```python
w2 = torch.rand(hidden_size, 
                output_size, 
                
                requires_grad=True)
w2.shape
```

```python
loss_func = torch.nn.CrossEntropyLoss()
```

```python
y_pred = 1/(1+(e**-(x.mm(w1).clamp(min=0).mm(w2))))
    
y_pred.size()
```

```python
y_pred
```

```python
y.size()
```

```python
y
```

```python
for i in range(1, 201):
    
    y_pred = 1/(1+(e**-(x.mm(w1).clamp(min=0).mm(w2))))
    loss = loss_func(y_pred,y)
    
    if i % 50 ==0:
        print(i, loss.item())
        
    loss.backward()
    
    with torch.no_grad():
        w1 -= learning_rate * w1.grad
        w2 -= learning_rate * w2.grad
        w1.grad.zero_()
        w2.grad.zero_()
```

```python
print ('w1: ', w1)
print ('w2: ', w2)
```

```python
class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(n_feature, n_hidden)   # hidden layer
        self.out = torch.nn.Linear(n_hidden, n_output)   # output layer

    def forward(self, x):
        x = F.relu(self.hidden(x))      # activation function for hidden layer
        x = self.out(x)
        return x
```

```python
net = Net(n_feature=2, n_hidden=10, n_output=2)     # define the network
print(net)
```

```python
optimizer = torch.optim.SGD(net.parameters(), lr=0.02)
loss_func = torch.nn.CrossEntropyLoss()
```

```python
for t in range(10):
    out = net(x)                 # input x and predict based on x
    loss = loss_func(out, y)     # must be (1. nn output, 2. target), the target label is NOT one-hotted

    optimizer.zero_grad()   # clear gradients for next train
    loss.backward()
    with torch.no_grad():
        w1 -= learning_rate * w1.grad
        w2 -= learning_rate * w2.grad
        
    optimizer.step()        # apply gradients

    if t % 2 == 0:
        # plot and show learning process
        plt.cla()
        prediction = torch.max(out, 1)[1]
        pred_y = prediction.data.numpy()
        target_y = y.data.numpy()
        plt.scatter(x.data.numpy()[:, 0], x.data.numpy()[:, 1], c=pred_y, s=100, lw=0, cmap='plasma')
        accuracy = float((pred_y == target_y).astype(int).sum()) / float(target_y.size)
        plt.text(1.5, -4, 'Accuracy=%.2f' % accuracy, fontdict={'size': 10, 'color':  'black'})
        plt.pause(0.1)
```
